# EXPERIMENTS

```julia
using JuLS
using Plots
```

## 1. Knapsack

```julia
# Instance
e = KnapsackExperiment("data/knapsack/ks_19_0")

# Model
model = init_model(e)

# Optimization
optimize!(model, limit=IterationLimit(100))
```

```julia
plot(-model.run_metrics.objective, legend=false, xlabel = "Iteration", ylabel = "Objective")
```

## 2.Travelling Salesman Problem

```julia
# Instance
e = TSPExperiment("data/knapsack/ks_19_0")

# Heuristics
init = SimpleInitialization()
pick = SimulatedAnnealing()

# Model
model = init_model(e, init = init, pick = pick)

# Optimization
optimize!(model, limit=IterationLimit(1000))
```

```julia
plot(model.run_metrics.objective, legend=false, xlabel = "Iteration", ylabel = "Objective")
```

## 3.Graph Coloring problem

```julia
# Instance
e = GraphColoringExperiment("data/graph_coloring/gc_20_1", 5)

# Heuristics
init = SimpleInitialization()
neigh = RandomNeighbourhood(20, 1)
pick = SimulatedAnnealing()

# Model
model = init_model(e, init = init, neigh = neigh, pick = pick)

# Optimization
optimize!(model, limit=IterationLimit(100))
```

```julia
plot(model.run_metrics.objective, legend=false, xlabel = "Iteration", ylabel = "Objective")
```
